```python
#Counting Sort

# counting sort will perform the sorting from the main array to the sorted array by using the formula of :-
# out[count[Num[i]]-1] = Num[i]
# count[Num[i]] = count[Num[i]] -1

def count_sort(Num): #define the count sort 
  n = len(Num) # find the length of the array
  out = [0]*n # setting the intial postion postion with index [0] for the output and it will be equal to the length of main array Num 
  count = [0]*10 # count will measure the comulative frequency of the array therefor we mark the index [0] and set it at 10 by default
  for i in range(0 , n):
    count[Num[i]] = count[Num[i]] + 1 # we can also write this as count[Num[i]] += 1 , we apllying this logic to count the element in main array 
  for i in range(1 , 10):
    count[i] = count[i] + count[i - 1] # here we try to measure the comulative frequncy where we try to add the current element with the element of previous index
  i = n -1  # represent the last element of the array 

  while i >= 0:
    out[count[Num[i]]-1] = Num[i]
    count[Num[i]] = count[Num[i]] -1
    i = i -1
  return out

    
```

```python
#taking the input from user 

Num = list(map(int , input().split()))

print(count_sort(Num))
```
